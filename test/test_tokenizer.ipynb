```python
from pathlib import Path
import json

from transformers import PreTrainedTokenizerFast
```

```python
CONFIG_DIR = Path("../configs")
TOKENIZER_FILE = CONFIG_DIR / "tokenizer.json"
TOKENIZER_CONFIG_FILE = CONFIG_DIR / "tokenizer_config.json"
SPECIAL_TOKENS_FILE = CONFIG_DIR / "special_tokens_map.json"

for path in [TOKENIZER_FILE, TOKENIZER_CONFIG_FILE, SPECIAL_TOKENS_FILE]:
    if not path.is_file():
        raise FileNotFoundError(f"Required file not found: {path}")

with TOKENIZER_CONFIG_FILE.open("r", encoding="utf-8") as f:
    tokenizer_config = json.load(f)

with SPECIAL_TOKENS_FILE.open("r", encoding="utf-8") as f:
    special_tokens_map = json.load(f)

print("Tokenizer files loaded successfully.")
print(f"- tokenizer file: {TOKENIZER_FILE}")
print(f"- tokenizer config: {TOKENIZER_CONFIG_FILE}")
print(f"- special tokens map: {SPECIAL_TOKENS_FILE}")
```

```python
tokenizer_kwargs = {
    "tokenizer_file": str(TOKENIZER_FILE),
    "padding_side": tokenizer_config.get("padding_side", "right"),
    "truncation_side": tokenizer_config.get("truncation_side", "right"),
}

for key in ["unk_token", "pad_token", "eos_token", "bos_token", "cls_token", "sep_token", "mask_token"]:
    value = special_tokens_map.get(key)
    if value is not None:
        tokenizer_kwargs[key] = value

tokenizer = PreTrainedTokenizerFast(**tokenizer_kwargs)
print("Tokenizer initialized successfully.")
```

```python
summary = {
    "vocab_size": tokenizer.vocab_size,
    "model_max_length": tokenizer.model_max_length,
    "padding_side": tokenizer.padding_side,
    "unk_token": tokenizer.unk_token,
    "unk_token_id": tokenizer.unk_token_id,
    "pad_token": tokenizer.pad_token,
    "pad_token_id": tokenizer.pad_token_id,
    "eos_token": tokenizer.eos_token,
    "eos_token_id": tokenizer.eos_token_id,
    "bos_token": tokenizer.bos_token,
    "bos_token_id": tokenizer.bos_token_id,
}

summary
```

```python
text = "RNA, Ambystoma mexicanum, LOC138478548, 5S ribosomal RNA<seq><gene><rRNA>GCCTACGGCCACACCACACTGAATGCGCCCGATCTCGTCTGATCTCGGAAGCTAAGCAGTGTCGGGCCTAGTTAGTACCTGGATGGGAGACCGCCTGGGAATACCAGGTGCTGTAGACT</rRNA></gene></seq><eos>"
encoded = tokenizer(text, return_tensors="pt", add_special_tokens=True)
encoded
```

```python
decoded = tokenizer.decode(encoded["input_ids"][0], skip_special_tokens=False)
print(decoded)
```
